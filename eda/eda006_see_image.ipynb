```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
import multiprocessing
```

```python
from glob import glob
```

```python
img_path = glob('/data/train/*jpg')
```

```python
path = img_path[0]
img = cv2.imread(path)
```

```python
(img[:,:,0]==img[:,:,1]).all()
```

```python
img_size = 640
img = cv2.resize(img, (img_size, img_size))
```

```python
img_1ch = img[:, :, 0]
```

```python
clahe = cv2.createCLAHE(clipLimit=10.0, tileGridSize=(8,8))
cl1 = clahe.apply(img_1ch)

plt.gray()
plt.figure(figsize=(20, 20))
plt.subplot(1, 2, 1)
plt.imshow(img_1ch)

plt.subplot(1, 2, 2)
plt.imshow(cl1)

plt.show()
```

```python
clahe = cv2.createCLAHE(clipLimit=10.0, tileGridSize=(16, 16))
cl1 = clahe.apply(img_1ch)

plt.gray()
plt.figure(figsize=(20, 20))
plt.subplot(1, 2, 1)
plt.imshow(img_1ch)

plt.subplot(1, 2, 2)
plt.imshow(cl1)

plt.show()
```

```python
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
cl1 = clahe.apply(img_1ch)

plt.gray()
plt.figure(figsize=(20, 20))
plt.subplot(1, 2, 1)
plt.imshow(img_1ch)

plt.subplot(1, 2, 2)
plt.imshow(cl1)

plt.show()
```

```python
def convert(p):
    img = cv2.imread(p)
    img = cv2.resize(img, (640, 640))
    s_p = p.replace("train", "train512")
    cv2.imwrite(s_p, img)
```

```python
pool = multiprocessing.Pool(16)
```

```python
pool.map(convert, img_path)
```
