```python
import ast
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt

from sklearn.metrics import roc_auc_score
```

```python
import math
def sigmoid(a):
    s = 1 / (1 + math.e**-a)
    return s
```

```python
df = pd.read_csv('/data/df_train2.csv')
df_anot = pd.read_csv('/data/train_annotations.csv')
df_bifurcation = pd.read_csv('/data/additional/RANZCR_CLiP_tracheal_bifurcation.csv')
```

```python
oof_800 = sigmoid(np.load('/workspace/output/exp024_test800_2/oof/oof.npy'))
oof_864 = sigmoid(np.load('/workspace/output/exp024_test864_2/oof/oof.npy'))
oof = (oof_800 + oof_864) / 2
```

```python
cvc_w_anot_index = df[df.StudyInstanceUID.isin(df_anot[df_anot.label.str.contains('CVC')].StudyInstanceUID.unique())].StudyInstanceUID.unique()
```

```python
pred_list = oof_800
cols = [
    'ETT - Abnormal', 'ETT - Borderline',
    'ETT - Normal', 'NGT - Abnormal', 'NGT - Borderline',
    'NGT - Incompletely Imaged', 'NGT - Normal', 'CVC - Abnormal',
    'CVC - Borderline', 'CVC - Normal', 'Swan Ganz Catheter Present'
]
targets_list = df.loc[:, cols].values


final_each_roc = []

for cv in range(5):
    each_roc = []
    val_index = df[df.cv == cv].index
    for i in range(11):
        each_roc.append(roc_auc_score(targets_list[val_index, i], pred_list[val_index, i]))
    mean_roc = np.mean(each_roc)
    final_each_roc.append(each_roc)

    print('-' * 20)
#     print(f'Oof score: {mean_roc}')
#     print(f'ETT_Abnormal score: {each_roc[0]}')
#     print(f'ETT_Borderline score: {each_roc[1]}')
#     print(f'ETT_Normal score: {each_roc[2]}')
#     print(f'NGT_Abnormal score: {each_roc[3]}')
#     print(f'NGT_Borderline score: {each_roc[4]}')
#     print(f'NGT_IncompletelyImaged score: {each_roc[5]}')
#     print(f'NGT_Normal score: {each_roc[6]}')
    print(f'CVC_Abnormal score: {each_roc[7]}')
    print(f'CVC_Borderline score: {each_roc[8]}')
    print(f'CVC_Normal score: {each_roc[9]}')
#     print(f'SwanGanzCatheterPresent score: {each_roc[10]}')
    print('-' * 20)
    
each_roc = np.mean(final_each_roc, axis=0)
print('+++++++++++++++++++++++++++++++++++++++++++++++++++++++')
print('-' * 20)
# print(f'Oof score: {mean_roc}')
# print(f'ETT_Abnormal score: {each_roc[0]}')
# print(f'ETT_Borderline score: {each_roc[1]}')
# print(f'ETT_Normal score: {each_roc[2]}')
# print(f'NGT_Abnormal score: {each_roc[3]}')
# print(f'NGT_Borderline score: {each_roc[4]}')
# print(f'NGT_IncompletelyImaged score: {each_roc[5]}')
# print(f'NGT_Normal score: {each_roc[6]}')
print(f'CVC_Abnormal score: {each_roc[7]}')
print(f'CVC_Borderline score: {each_roc[8]}')
print(f'CVC_Normal score: {each_roc[9]}')
# print(f'SwanGanzCatheterPresent score: {each_roc[10]}')
print('-' * 20)
```
